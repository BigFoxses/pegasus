```python
import boto.ec2
import argparse
import json
from pandas import DataFrame
```

```python
def get_all_ec2_instances():
    regions = ["ap-northeast-1", "ap-southeast-1", "ap-southeast-2", 
               "eu-central-1", "eu-west-1", "sa-east-1", 
               "us-east-1", "us-west-1", "us-west-2"]
    
    inst_info = {'groups': [],
                 'public_dns_name': [],
                 'private_dns_name': [],
                 'key_name': [],
                 'instance_type': [],
                 'placement': [],
                 }
    
    for region in regions:
        ec2_conn = boto.ec2.connect_to_region(region)
        reservations = ec2_conn.get_all_reservations()
    
    
        for reservation in reservations:
            instances = reservation.instances

            for inst in instances:
                inst_info['public_dns_name'].append(inst.public_dns_name)
                inst_info['private_dns_name'].append(inst.private_dns_name)
                inst_info['key_name'].append(inst.key_name)
                inst_info['instance_type'].append(inst.instance_type)
                inst_info['groups'].append(inst.groups)
                inst_info['placement'].append(inst.placement)            
            
    return inst_info
```

```python
df = DataFrame(get_all_ec2_instances())
```

```python
instances_per_type = df[["instance_type", "groups"]].groupby(["instance_type"]).count()
instances_per_type.columns = ["num_instances"]
instances_per_type
```

```python
instances_per_key = df[["key_name", "groups"]].groupby(["key_name"]).count()
instances_per_key.columns = ["num_instances"]
instances_per_key
```

```python
instances_per_region = df[["placement", "groups"]].groupby(["placement"]).count()
instances_per_region.columns = ["num_instances"]
instances_per_region
```
